```python
import numpy as np
import pandas as pd
```

```python
def featureNormalise(x1,x2):
    m1 = x1.mean()
    m2 = x2.mean()
    std1 = x1.std()
    std2 = x2.std()
    MeanX = np.zeros([m,2])
    STDX = np.zeros([m,2])
    for i in range(m):
        MeanX[i][0] = x1[i] - m1
        MeanX[i][1] = x2[i] - m2
        STDX[i][0] = x1[i]/std1
        STDX[i][1] = x2[i]/std2
    return [MeanX, STDX]
```

```python
df = pd.read_csv('ex1data2.txt')
X = pd.DataFrame(df.iloc[:,0:2])
y = pd.DataFrame(df.iloc[:,2])
m = len(y)
```

```python
Mean, Std = featureNormalise(X['area'], X['rooms'])
print(Mean)
print()
print(Std)
```
