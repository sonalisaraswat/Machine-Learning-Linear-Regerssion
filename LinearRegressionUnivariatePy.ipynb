```python
import pandas as pd
import numpy as np
```

```python
def ComputeCost(X, y, theta):
    J = (1/(2*m))*sum((X.dot(theta)[0]-y)**2)
    return J
```

```python
def GradientDecent(X, y, theta, alpha, itera):
    x= pd.DataFrame(X.iloc[:,1])
    for i in range(0,itera):
        t1 = theta[0,:] - (alpha/m)*sum(X.dot(theta)[0] - y)
        t2 = theta[1,:] - (alpha/m)*(x.T).dot(X.dot(theta)[0] - y)
        theta[0,:] = t1
        theta[1,:] = t2
    return theta
```

```python
df = pd.read_csv('ex1data1.txt')
X = pd.DataFrame(df.iloc[:,0])   # X as a data frame
y = df.iloc[:,1]                 # y as a series
```

```python
m = len(y)
alpha = 0.01
itera = 1500
Extra = np.ones([m,1], dtype = int)
X = pd.concat([pd.DataFrame(Extra), X], axis = 1)
theta = np.array([[0],[0]], dtype = float)
```

```python
print(ComputeCost(X, y, theta))   # without Gradient decent application
theta = GradientDecent(X,y,theta,alpha,itera)  #appropriate values of theta
print(ComputeCost(X,y,theta))   # by getting better theta by gradient decent
```

```python
#Given profit = 35,000 i.e 3.5....predict profit
input = np.array([1, 3.5])
predict = (input.dot(theta))*10000
print(predict)
```

```python
#Given profit = 70,000 i.e 7....predict profit
input = np.array([1, 7])
predict = (input.dot(theta))*10000
print(predict)
```
